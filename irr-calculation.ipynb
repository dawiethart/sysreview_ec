```python
intersecting_titles = set(df_d['Autor, Herausgeber oder Institution']).intersection(set(df_l['Autor, Herausgeber oder Institution']))

columns =  ['Autor, Herausgeber oder Institution', 'pos. findings ec hypothesis', 'neg. findings ec hypothesis','effect: polarization', 'effect: misinformation', 'effect:trust loss', 'effect: extremism', 'effect: no effect', 'antecedent: recommender system', 'antecedent and property: user behaviour', 'antecedent: offline fragmentation ', 'antecedent: institutions and other media sources', 'effect: populism', 'property: overstated', 'property: event-specific',  'property: active users and efficacy', 'effect: toxicity', 'antecedent and property: group behaviour']

df_filtered_d = df_d[df_d['Autor, Herausgeber oder Institution'].isin(intersecting_titles)][columns]
df_filtered_l = df_l[df_l['Autor, Herausgeber oder Institution'].isin(intersecting_titles)][columns]
#df_filtered_d[columns] = df_filtered_d.fillna(0).astype(int)
#df_filtered_l[columns] = df_filtered_l.fillna(0).astype(int)
# sort by author /title 
df_filtered_d = df_filtered_d.sort_values(by='Autor, Herausgeber oder Institution')
df_filtered_l = df_filtered_l.sort_values(by='Autor, Herausgeber oder Institution')

new_columns = ['pos. findings ec hypothesis', 'neg. findings ec hypothesis', 'effect: polarization', 'effect: misinformation', 'effect:trust loss', 'effect: extremism', 'effect: no effect', 'antecedent: recommender system', 'antecedent and property: user behaviour', 'antecedent: offline fragmentation ', 'antecedent: institutions and other media sources', 'effect: populism', 'property: overstated', 'property: event-specific', 'property: active users and efficacy', 'effect: toxicity', 'antecedent and property: group behaviour']

df_filtered_d = df_filtered_d[new_columns].fillna(0).astype(int)
df_filtered_l= df_filtered_l[new_columns].fillna(0).astype(int)

```

```python
#to numpy and flatten
df_filtered_d = df_filtered_d.to_numpy().flatten()
df_filtered_l = df_filtered_l.to_numpy().flatten()
```

```python
#calculate interrater reliability
from sklearn.metrics import cohen_kappa_score
import numpy as np
cohen_kappa_score(df_filtered_d, df_filtered_l)


```
